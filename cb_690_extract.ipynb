```python
layer_0 = nib.load("data/cb690/t1_t2/layer_000_cd8f8368264427396b9230d43a99f8f8.nii")
layer_1 = nib.load("data/cb690/t1_t2/layer_001_0f92348613a06b55e36d38584a63df43.nii")
layer_4 = nib.load("data/cb690/t1_t2/layer_004_3fbf1eb3e2369cc865ef2b690e17c794.nii")
img = layer_4
data = img.get_fdata()
voxel_vol = np.prod(img.header.get_zooms()[:3])
```
